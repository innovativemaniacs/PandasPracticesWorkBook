```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
import numpy as np
```

```python
df=pd.read_csv("kkk.csv")
df
```

```python
df=df[(df["Year"]>=1947) & (df["Year"]<=2024)]
df
```

```python
new_filtered_df=df.groupby(["Year", "Cause"]).size().reset_index(name="count").sort_values("count", ascending=False)
new_filtered_df

df_Year_between_1947_to2024=new_filtered_df[(new_filtered_df["Year"]>=1947) & (new_filtered_df["Year"]<=2025)]
df_Year_between_1947_to2024

```

```python
# yearwise casualties
casualties_by_year = df_Year_between_1947_to2024.groupby("Year")["count"].sum().reset_index()
casualties_by_year




```

```python


plt.figure(figsize=(16, 6))
sns.lineplot(data=casualties_by_year, x="Year", y="count", errorbar=None)

plt.title("Trend of Total Casualties from 1947 to 2024")
plt.xlabel("Year")
plt.ylabel("Number of Casualties")
plt.xticks(rotation=90)
plt.grid(True)
plt.tight_layout()
plt.show()
```

```python
plt.figure(figsize=(25, 12))
ax = sns.barplot(data=casualties_by_year, x="Year", y="count", errorbar=None)


for container in ax.containers:
    ax.bar_label(container, fontsize=8, rotation=90)

plt.title("Trend of Total Casualties from 1947 to 2024")
plt.xlabel("Year")
plt.ylabel("Number of Casualties")
plt.xticks(rotation=90)
plt.grid(True)
plt.tight_layout()
plt.show()
```
